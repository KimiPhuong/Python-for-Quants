# Rows & Columns

```python
def makeDF():
    d = np.arange(16).reshape(4,4)
    cols = ['one', 'two', 'three', 'four']
    ndx = ['a', 'b', 'c', 'd']

    df = pd.DataFrame(data=d, columns=cols, index=ndx)

    return df
```

```python
df = makeDF()

df
```
