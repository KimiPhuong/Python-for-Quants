# Resampling

- Down Sample - V V Easy
- Up Sample - Easy

## Down Sampling

Fine Grained to Coarse Grained time

```python
df.resample(rule='Y').mean()
df.resample(rule='Q').mean()
df.loc['2017'].resample(rule='M').mean()
```

## Upsampling

```python
df2 = df.loc['2017'].resample(rule='M').mean()

df2.resample(rule='W').mean()
df2.resample(rule='W').ffill(limit=2)
df2.resample(rule='W').bfill()
df2.resample(rule='D').interpolate()
```

```python
df_interp = pd.DataFrame()

df_interp['LIN'] = df2['Adj Close'].resample(rule='D').interpolate(method='linear')
df_interp['QUAD'] = df2['Adj Close'].resample(rule='D').interpolate(method='quadratic')
df_interp['CUBIC'] = df2['Adj Close'].resample(rule='D').interpolate(method='cubic')
```

```python
df_interp.plot(figsize=(18, 6))
```

# Moving Windows

```python
df = yf.download(tickers='^GSPC', start='2000-01-01')

df['Adj Close'].plot()
df['Adj Close'].rolling(window=42).mean().plot()
df['Adj Close'].rolling(window=252).mean().plot()
```

```python
df['Adj Close'].rolling(window=42).mean().plot()
df['Adj Close'].rolling(window=252).mean().plot()
```

```python
df['Adj Close'].rolling(window=252).max().plot()
df['Adj Close'].rolling(window=252).min().plot()
```

```python
df['Adj Close'].rolling(window=252).mean().plot()
df['Adj Close'].expanding().mean().plot()
```

```python
df['Adj Close'].rolling(window=252).mean().plot()
df['Adj Close'].ewm(span=252).mean().plot()
```
