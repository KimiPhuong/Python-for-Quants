```python
import pandas as pd
import numpy as np
```

```python
df_Basket = pd.read_excel(io='../Data/regression.xlsx', index_col = 'Date',sheet_name ='Basket')
df_PXD = pd.read_excel(io='../Data/regression.xlsx', index_col = 'Date',sheet_name ='PXD')
df_Basket.head()
df_PXD.head()
```

```python
df_Basket = pd.read_excel(io='../Data/regression.xlsx', index_col = 'Date',sheet_name ='Basket').pct_change()
df_PXD = pd.read_excel(io='../Data/regression.xlsx', index_col = 'Date',sheet_name ='PXD').pct_change()
```

```python
df_Returns = pd.concat(objs=[df_PXD,df_Basket],axis=1)
df_Returns.head()
```

```python
import statsmodels.formula.api as sm
model = sm.ols(formula = 'PXD ~ SPY + Oil + Gold + NG',data = df_Returns)
results = model.fit()
results.summary()
```

```python
def regress(symbol,df):
    formula = symbol + '~ SPY + Oil + Gold + NG'
    model = sm.ols(formula = formula ,data = df)
    return  model.fit()
```

```python
report = regress('PXD',df_Returns)
report.summary()
fltr = report.pvalues < 0.05
report.params[fltr]
```

```python
def regress(symbol,df):
    formula = symbol + '~ SPY + Oil + Gold + NG'
    result = sm.ols(formula = formula ,data = df).fit()
    fltr = result.pvalues < 0.05
    df = pd.DataFrame(data=result.params[fltr],columns=[symbol])
    return df
regress('PXD',df_Returns)
```

```python
df_Stocks = pd.read_excel(io='../Data/regression.xlsx', index_col = 'Date',sheet_name ='StockPrices').pct_change()
df_Stocks.head()
```

```python
df_Returns = pd.concat(objs=[df_Stocks,df_Basket], axis=1)
```

```python
df_Returns.describe()
```

```python
df_Beta = pd.DataFrame()

for col in df_Returns.columns[:-4]:
    df = regress(col,df_Returns)
    df_Beta = pd.concat([df_Beta,df],axis=1,sort=False)
print('Done')
```

```python
writer = pd.ExcelWriter(path='./Betas.xlsx')
df_Beta.to_excel(excel_writer=writer,sheet_name='Betas')
df_Beta.transpose().to_excel(excel_writer=writer,sheet_name='Tposed')
writer.save()
```
